```python
temp = df.dropna(subset=['CustomerID'])
temp.isna().sum()
```

```python
temp['total_price'] = temp['UnitPrice']*temp['Quantity']
temp.head()
```

```python
cleanDF = temp.copy()
```

```python
Monetary = cleanDF[['CustomerID', 'total_price']].groupby('CustomerID').agg('sum').reset_index()
Monetary.head()
```

```python
Recency = cleanDF[['CustomerID', 'InvoiceDate']].groupby('CustomerID').agg('max').reset_index()
Recency.head()
```

```python
Frequency = cleanDF[['CustomerID', 'InvoiceNo']].groupby('CustomerID').agg('nunique').reset_index()
Frequency.head()
```
